PLOTTING DATA ACCORDING TO THE STARS GIVEN BY AMAZON USERS

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

DATA FROM PAGES 1-5

Opens the file for reading the data

```python
with open('C:\/items.json') as f:
  data = json.load(f)
```

```python
# print(len(data['product_name']))
# print(len(data['product_review']))
```

Retreives the data from the data and stores them in respective arrays

```python
names=[] #name of the product
ratings=[] #rating of the product
price=[] #price of the product
left=[] #the product serial number
count=0

for j in range(0,len(data),1):
    data1=data[j]
#     print(len(data1['product_name']))
#     print(len(data1['product_review']))
    for i in range(0,len(data1['product_review'])):
        count+=1
        left.append(count)
        names.append(data1['product_name'][i])
        price.append(data1['product_price'][i])
        ls=data1['product_review'][i].split()
        ratings.append(ls[0])
```

Displaying the analysis in terms of ratings 

```python
x=np.array(left)
ratingsfloat=[]
for i in ratings:
    ratingsfloat.append(float(i))
y=np.array(ratingsfloat)
plt.xticks(np.arange(0, len(names)+1, 1),fontsize=20)
plt.yticks(np.arange(0, len(ratings)+1, 0.5),fontsize=18)
plt.bar(x,y)
plt.rcParams["figure.figsize"] = (50,5)
plt.ylim(ymin=0) 
plt.show()
for i in range(0,len(left),1):
    print(left[i],"\t",names[i],"\t",ratings[i])
```
