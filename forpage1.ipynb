```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

opening the json file in notebook

```python
with open('C:\/items_first.json') as f:
  data = json.load(f)
data=data[0]
```

retriving prices of the product

```python
pp=[]
for i in range(0,len(data['product_price']),1):
    if(data['product_price'][i]!='.'):
        pp.append(data['product_price'][i])  
print(len(pp))
```

retriving the other datas of the product

```python
count=0
left=[]
names=[]
price=[]
ratings=[]
for i in range(0,len(data['product_name'])):
    count+=1
    left.append(count)
    names.append(data['product_name'][i])
    price.append(pp[i])
    ls=data['product_review'][i].split()
    ratings.append(ls[0])
ratings_price=[]
for i in range(0,len(price),1):
    ratings_price.append(round((float(price[i])/float(ratings[i])),3))
```

displaying the arrays where the data is stored

```python
print(len(left))
print(names)
print(len(price))
print(ratings)
print(ratings_price)
```

analysis for book in the format : ratings/prices of each book

```python
x=np.array(left)
y=np.array(ratings_price)
plt.xticks(np.arange(0, len(names)+1, 1),fontsize=20)
plt.yticks(np.arange(0, len(ratings_price)+1, 0.5),fontsize=18)
plt.bar(x,y)
plt.rcParams["figure.figsize"] = (50,10)
plt.ylim(ymin=0) 
plt.show()
for i in range(0,len(left),1):
    print(left[i],"\t",names[i],"\t",ratings_price[i])
```
